```python
#A dictionary (created) mapped gesture recognising model, still have to be added to other modules
import cv2
import mediapipe as mp

# Initialize MediaPipe Hand module
mp_drawing = mp.solutions.drawing_utils
mp_hands = mp.solutions.hands

# Initialize OpenCV VideoCapture
cap = cv2.VideoCapture(0)  # You can change the argument to use a different camera or video file

# Create a dictionary to map hand landmarks to gestures
gestures = {
    "Fist": [0, 4, 8, 12, 16, 20],# Valid indices are 0 to 20
    "Open Hand": [0, 5, 9, 13, 17],
    "Thumbs Up": [1, 2],
    "Victory Sign": [5, 9],
    "Pointing Finger": [8],
    "Rock 'n' Roll": [3, 5, 9],
    "Peace Sign": [6, 10],
    "Thumb and Pinky Out": [1, 20],
    "Three Fingers": [6, 10, 14],
    "OK Sign": [0, 8]
}
with mp_hands.Hands(min_detection_confidence=0.5, min_tracking_confidence=0.5) as hands:
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            continue

        frame = cv2.flip(frame, 1)
        height, width, _ = frame.shape

        # Convert the BGR image to RGB
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

        # Process the image with MediaPipe Hand module
        results = hands.process(frame_rgb)

        # Check if any hands were detected
        if results.multi_hand_landmarks:
            for hand_landmarks in results.multi_hand_landmarks:
                # Draw landmarks on the frame
                mp_drawing.draw_landmarks(frame, hand_landmarks, mp_hands.HAND_CONNECTIONS)

                # Detect gestures
                detected_gesture = None
                landmarks = hand_landmarks.landmark
                for gesture, key_points in gestures.items():
                    try:
                        if all(0 <= i < 21 for i in key_points):
                            is_gesture = all(
                                landmarks[i].y < landmarks[i + 1].y for i in key_points if i < 21
                            )
                            # Add more specific checks here to determine if the gesture is being performed correctly.
                            if is_gesture:
                                detected_gesture = gesture
                    except IndexError:
                        # Handle the IndexError exception
                        print("IndexError: list index out of range")

                # Display the detected gesture
                if detected_gesture:
                    print(detected_gesture)

                    cv2.putText(
                        frame,
                        detected_gesture,
                        (50, 50),
                        cv2.FONT_HERSHEY_SIMPLEX,
                        1,
                        (0, 0, 255),
                        2,
                        cv2.LINE_AA,
                    )

        # Display the frame
        cv2.imshow("Hand Gesture Recognition", frame)

        # Quit if 'Esc' key is pressed
        if cv2.waitKey(1) & 0xFF == 27:
            break

# Release the camera and destroy all windows
cap.release()
cv2.destroyAllWindows()
```
